```python
# Asumiendo que ya tienes los DataFrames cargados: df_buses, df_colectivos, df_metros

# Función para limpiar y validar un DataFrame
def limpiar_validar_df(df, nombre_df):
    print(f"Limpieza y validación de {nombre_df}:\n")

    # 1. Verificar y eliminar duplicados
    antes = df.shape[0]
    df.drop_duplicates(inplace=True)
    despues = df.shape[0]
    print(f"Eliminados {antes - despues} duplicados. Total de filas ahora: {despues}\n")

    # 2. Verificar y manejar valores nulos
    nulos = df.isnull().sum()
    print(f"Valores nulos por columna:\n{nulos}\n")

    # Opcional: manejar nulos (esto dependerá de tu contexto, podrías querer eliminar filas o rellenar con algún valor)
    # Ejemplo: eliminar filas con nulos
    df.dropna(inplace=True)
    despues_nulos = df.shape[0]
    print(f"Eliminados {despues - despues_nulos} filas con valores nulos. Total de filas ahora: {despues_nulos}\n")

    # 3. Validar los rangos de valores (esto también es solo un ejemplo)

    if 'latitud' in df.columns and 'longitud' in df.columns:
        fuera_rango = df[(df['latitud'] < -33) | (df['latitud'] >-33 ) | (df['longitud'] < -70) | (df['longitud'] > -70)].shape[0]
        print(f"Filas con coordenadas fuera de rango: {fuera_rango}\n")

    print(f"Limpieza y validación de {nombre_df} completa.\n")
    return df

# Aplicar la limpieza y validación a cada DataFrame
df_buses = limpiar_validar_df(df_buses, 'df_buses')
df_colectivos = limpiar_validar_df(df_colectivos, 'df_colectivos')
df_metro = limpiar_validar_df(df_metro, 'df_metros')

# Ver los primeros registros después de la limpieza
print("Primeros registros de df_buses:")
print(df_buses.head())
print("\nPrimeros registros de df_colectivos:")
print(df_colectivos.head())
print("\nPrimeros registros de df_metros:")
print(df_metro.head())
```

**4. Análisis de Satisfacción del Usuario**

```python
# Datos de satisfacción
satisfaccion = pd.DataFrame({
    'Criterio': ['Puntualidad', 'Comodidad', 'Seguridad'],
    'Colectivos': [5, 5, 5],
    'Buses': [8, 4, 6],
    'Metro': [9, 7, 9]
})

# Calcular porcentajes relativos basados en una escala máxima de 10
satisfaccion_porcentajes = satisfaccion.copy()

for col in satisfaccion.columns[1:]:
    satisfaccion_porcentajes[col] = (satisfaccion[col] / 10) * 100

# Preparar datos para el gráfico de barras agrupadas
barWidth = 0.25
r1 = range(len(satisfaccion))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]

plt.figure(figsize=(10, 6))

plt.bar(r1, satisfaccion_porcentajes['Colectivos'], color='b', width=barWidth, edgecolor='grey', label='Colectivos')
plt.bar(r2, satisfaccion_porcentajes['Buses'], color='g', width=barWidth, edgecolor='grey', label='Buses')
plt.bar(r3, satisfaccion_porcentajes['Metro'], color='r', width=barWidth, edgecolor='grey', label='Metro')

plt.xlabel('Criterio', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(satisfaccion))], satisfaccion['Criterio'])
plt.ylabel('Porcentaje de Satisfacción')
plt.title('Satisfacción del Usuario por Criterio y Tipo de Transporte')

plt.legend()
plt.show()


```

**5. Frecuencia de servicios por días de la semana**

```python
# Datos de horas de servicio por día
frecuencia_df = pd.DataFrame({
    'Día de la Semana': ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo'],
    'Colectivos': [75, 75, 75, 75, 70, 75, 75],
    'Buses': [100, 100, 100, 100, 95, 100, 100],
    'Metro': [135, 130, 130, 130, 125, 130, 130]
})

# Promedios por tipo de transporte
promedio_metro = frecuencia_df['Metro'].mean()
promedio_buses = frecuencia_df['Buses'].mean()
promedio_colectivos = frecuencia_df['Colectivos'].mean()

# Plotear la frecuencia de servicios
ax = frecuencia_df.plot(x='Día de la Semana', kind='bar', figsize=(12, 6))

# Añadir títulos y etiquetas
plt.title('Paradas de Servicio por Día y Tipo de Transporte')
plt.xlabel('Día de la Semana')
plt.ylabel('Paradas de Servicio')


# Añadir líneas horizontales para destacar el promedio de horas de servicio
ax.axhline(y=promedio_metro, color='gray', linestyle='--', linewidth=0.8)
ax.axhline(y=promedio_buses, color='gray', linestyle='--', linewidth=0.8)
ax.axhline(y=promedio_colectivos, color='gray', linestyle='--', linewidth=0.8)

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

```python
# Crear histogramas de densidad de servicios
def crear_histograma(horarios, tipo_transporte):
    horarios_list = [int(h.split(':')[0]) for h in horarios.replace(' ', '').split('-')]
    plt.hist(horarios_list, bins=24, alpha=0.5, label=tipo_transporte)

# Histograma para colectivos
crear_histograma(df_colectivos.iloc[0]['HORARIO_SEMANA'], 'Colectivos')
# Histograma para buses
crear_histograma(df_buses.iloc[0]['horario_semana'], 'Buses')
# Histograma para metro
crear_histograma(df_metro.iloc[0]['horario_metro_semana'], 'Metro')

plt.title('Densidad de Servicios por Horarios')
plt.xlabel('Hora del Día')
plt.ylabel('Frecuencia de Servicios')
plt.legend()
plt.show()
```
